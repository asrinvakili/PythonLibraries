## Import Pandas:

```python
import pandas as pd
```

```python
a1 = [1, 2, 3, 4, 5]

pd1 = pd.Series(a1)

pd1
```

```python
a2 = [1, 2, 3, 4, 5]

b2 = ['a', 'b', 'c', 'd', 'e']

pd2 = pd.Series(b2, index = a2)

pd2
```

```python
a3 = {"BMW": 18000, "Volvo": 70000, "Ford": 20000}

pd3 = pd.Series(a3)

pd3
```

```python
data1 = {
  'cars': ["BMW", "Volvo", "Ford"],
  'price': [18000, 70000, 20000]
}

df1 = pd.DataFrame(data1)

df1
```

```python
#refer to the row index:

df1.loc[1]
```

```python
#use a list of indexes:

df1.loc[[0, 1]]
```

```python
data2 = {
  'cars': ["BMW", "Volvo", "Ford"],
  'price': [18000, 70000, 20000]
}

df2 = pd.DataFrame(data2, index = ['car1', 'car2', 'car3'])

df2.loc['car3']
```
